```python
import pandas as pd
import seaborn as sns
```

**Biblioteca seaborn**

Biblioteca construida em cima da Marplotlib, para facilitar a visualização de gráficos.

→ Funciona de uma maneira mais simplificada, e possui alguns datasets prontos para trabalhar e explorar.

```python
# Visualizar datasets disponiveis
sns.get_dataset_names()
```

```python

# Exemplo de uso
iris = sns.load_dataset('iris')

iris
```

Função `pairplot()`, vai pegar todas as variaveis númericas em uma matriz de gráficos de dispersão com histogramas, uma técnica bastante comum em análises exploratórias de dados (EDA). Esses gráficos são frequentemente usados para avaliar relações entre variáveis e verificar distribuições individuais

```python
sns.pairplot(iris, hue='species')
```

```python
# Exemplo de uso
titanic = sns.load_dataset('titanic')

sns.pairplot(titanic, hue='alive')
```
